# 1. Драка

```python
import random

class Warrior:
    
    def __init__(self, name, points=100):
        self.name = name
        self.points = points

user1 = Warrior('Воин1')
user2 = Warrior('Воин2')
i = 0
while user1.points > 0 and user2.points > 0:
    user = random.randint(1, 2)
    i += 1
    if user == 1:
        user1.points -= 20
    else:
        user2.points -=20 
    print(i, user1.name, user1.points, user2.name, user2.points)
```

# 3. Окружность

```python
import math

class Circle:
    
    def __init__(self, x=0, y=0, r=0):
        self.x = x
        self.y = y
        self.r = r
    
    def square(self): 
        return self.r ** 2 * math.pi
    
    def perimeter(self):
        return 2 * math.pi * self.r
        
    def increase(self, K):
        self.r *= K
    
    def cross(self, other):
        if math.sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2) <= (self.r + other.r):
            print('Окружности пересекаются')
        else:
            print('Окружности не пересекаются')   

circle1 = Circle(0, 0, 5)
circle2 = Circle(1, 2, 5)
circle1.increase(2)
print(circle1.r)
print(circle1.square())
circle1.cross(circle2)
```

# 4. Отцы, матери и дети

```python
class Parent:

    def __init__(self, name, age, child):
        self.name, self.age, self.child = name, age, child

    def inform(self):
        new_str = ', '.join(self.child)
        print(f'Меня зовут {self.name}, мне {self.age}, мои дети: {new_str}.')

    def calm(self, child):
        for elem in self.child:
            if elem == child.name:
                child.calmness = True

    def feed(self, child):
        for elem in self.child:
            if elem == child.name:
                child.fullness = True


class Child:

    def __init__(self, name, age, calmness, fullness):
        self.name, self.age, self.calmness, self.fullness = name, age, calmness, fullness

    def inform(self):
        print(f'Меня зовут {self.name}, мне {self.age},', end=' ')
        if self.calmness == True:
            print('спокоен, ', end='')
        else:
            print('не спокоен, ', end='')
        if self.fullness == True:
            print('сыт.')
        else:
            print('голоден.')


parent1 = Parent('Борис', 44, ['Ваня', 'Маша'])
parent2 = Parent('Серафима', 67, ['Борис'])
child1 = Child('Ваня', 4, True, False)
child2 = Child('Маша', 10, False, True)
parent2.inform()
parent1.calm(child2)
parent1.feed(child1)
child1.inform()
child2.inform()
```

# 5. Весёлая ферма 

```python
class Gardener:
    
    def __init__(self, name):
        self.name = name
    
    def pick(self, potato):
        print(potato.number, 'созрела.')
    
    def care(self, potato):
        potato.stage += 1
        print('"', potato.number, '", состояние: ', potato.stage)
        
class Potato:
    
    def __init__(self, number, stage=0):
        self.number, self.stage = number, stage

my_bed = []
chel = Gardener('Виталик')
for n in range(3):
    my_bed.append(Potato(n))
for potato in my_bed:
    if potato.stage == 4:
        chel.pick(potato)
    else:
        chel.care(potato)
```

# 6. Магия

```python
class Water:
    name = 'Вода'
    
    def __add__(self, other):
        if other.name == 'Воздух':
            return Storm()
        elif other.name == 'Огонь':
            return Steam()
        elif other.name == 'Земля':
            return Dirt()
        else:
            return None

class Air:
    name = 'Воздух'
    
    def __add__(self, other):
        if other.name == 'Огонь':
            return Light()
        elif other.name == 'Земля':
            return Dust()
        elif other.name == 'Вода':
            return Storm()
        else:
            return None

class Fire:
    name = 'Огонь'
    
    def __add__(self, other):
        if other.name == 'Вода':
            return Steam()
        elif other.name == 'Воздух':
            return Light()
        elif other.name == 'Земля':
            return Lava()
        else:
            return None
        
class Ground:
    name = 'Земля'
    
    def __add__(self, other):
        if other.name == 'Вода':
            return Dirt()
        elif other.name == 'Воздух':
            return Dust()
        elif other.name == 'Огонь':
            return Lava()
        else:
            return None
        
class Storm:
    name = 'Шторм'
    
class Steam:
    name = 'Пар'
    
class Dirt:
    name = 'Грязь'
    
class Light:
    name = 'Молния'
    
class Lava:
    name = 'Лава'
    
elem1 = Water()
elem2 = Air()
elem3 = Fire()
elem4 = elem1 + elem2
elem5 = elem2 + elem3
print(f'{elem1.name} + {elem2.name} = {elem4.name}')
print(f'{elem2.name} + {elem3.name} = {elem5.name}')
```

# 7. Совместное проживание

```python
import random

class House:
    food = 50
    money = 0
    
class Man:
    
    def __init__(self, name, health=50):
        self.name, self.health = name, health
    
    def eat(self):
        House.food -= 1
        self.health += 1
        
    def work(self):
        House.money += 1
        self.health -= 1
    
    def play(self):
        self.health -= 1
        
    def shop(self):
        House.money -= 1
        House.food += 1
        
person1, person2 = Man('Вася'), Man('Маша')
for day in range(1, 366):
    n = random.randint(1, 6)
    person = random.choice([person1, person2])
    if person.health < 0:
        print(f'{person.name} умер(ла)')
        break
    if person.health < 20 and House.food > 10:
        person.eat()
    elif House.food < 10:
        person.shop()
    elif House.money < 50:
        person.work()
    elif n == 1:
        person.work()
    elif n == 2:
        person.eat()
    else:
        person.play()    
```

# 8. Блек-джек

```python
import random


class Two:
    def __init__(self):
        self.name = '2'
        self.points = 2


class Three:
    def __init__(self):
        self.name = '3'
        self.points = 3


class Four:
    def __init__(self):
        self.name = '4'
        self.points = 4


class Five:
    def __init__(self):
        self.name = '5'
        self.points = 5


class Six:
    def __init__(self):
        self.name = '6'
        self.points = 6


class Seven:
    def __init__(self):
        self.name = '7'
        self.points = 7


class Eight:
    def __init__(self):
        self.name = '8'
        self.points = 8


class Nine:
    def __init__(self):
        self.name = '9'
        self.points = 9


class Ten:
    def __init__(self):
        self.name = '10'
        self.points = 10


class Jack:
    def __init__(self):
        self.name = 'Jack'
        self.points = 10


class Queen:
    def __init__(self):
        self.name = 'Queen'
        self.points = 10


class King:
    def __init__(self):
        self.name = 'King'
        self.points = 10


class Ace:
    def __init__(self, sum_score=21):
        self.name = 'Ace'
        if sum_score <= 10:
            self.points = 11
        else:
            self.points = 1


class BlackJackGame:
    def __init__(self):
        self.pack_of_cards = [[Two().name, Two().points], [Three().name, Three().points], [Four().name, Four().points],
                              [Five().name, Five().points], [Six().name, Six().points], [Seven().name, Seven().points],
                              [Eight().name, Eight().points], [Nine().name, Nine().points], [Ten().name, Ten().points],
                              [Jack().name, Jack().points], [Queen().name, Queen().points],
                              [King().name, King().points], [Ace().name, Ace().points]] * 4
        self.your_score = 0
        self.croupier_score = 0

    def random_card(self, score, who_plays):
        if len(self.pack_of_cards) != 0:
            card = random.choice(self.pack_of_cards)
            card_points = card[1]
            card_name = card[0]
            self.pack_of_cards.remove(card)
        else:
            raise IndexError
        if card_name == 'Ace':
            score += Ace(score).points
        else:
            score += card_points
        self.who_get_card(score, card_name, who_plays)
        return score

    @staticmethod
    def who_get_card(score, card_name, who_plays):
        if who_plays == 'H':
            print(f'Your random card is {card_name}. You have {score} points for it.')
        else:
            print(f"Croupier's random card is {card_name}. He has {score} points for it.")

    def game(self):
        your_score = 0
        croupier_score = 0
        print('Initial 2 random cards:')
        print('YOURS:')
        for i in range(2):
            print(i + 1, '-> ', end='')
            your_score += self.random_card(self.your_score, 'H')
        print('Your total score: ', your_score)
        print("CROUPIER'S:")
        for i in range(2):
            print(i + 1, '-> ', end='')
            croupier_score += self.random_card(self.croupier_score, 'C')
        print("Croupier's total score: ", croupier_score)
        while True:
            answer = int(input('Do you want to take a card? 1 - yes, 2 - no \n'))
            if answer == 1:
                your_score = self.random_card(your_score, 'H')
                if croupier_score < 19 and your_score <= 21:
                    croupier_score = self.random_card(croupier_score, 'C')
                if your_score > 21 or croupier_score == 21:
                    print('You lost :(')
                    break
                elif your_score == 21 and croupier_score == 21:
                    print('Draw.')
                    break
                elif your_score == 21 or croupier_score > 21:
                    print('You won :)')
                    break
            elif answer == 2:
                if your_score > croupier_score and croupier_score < 19:
                    while croupier_score < 19:
                        croupier_score = self.random_card(croupier_score, 'C')
                if your_score < croupier_score <= 21:
                    print(f'You lost, you have {your_score} points, croupier has {croupier_score} points.')
                else:
                    print(f'You won, you have {your_score} points, croupier has {croupier_score} points.')
                break

    def start_game(self):
        random.shuffle(self.pack_of_cards)
        print('Game has started.')
        self.game()
```

# 9. Крестики-нолики

```python
class Player:
    def __init__(self, name='x'):
        self.name = name

class Game:
    def __init__(self):
        self.board = [[0 for i in range(3)] for j in range(3)]
        self.board_to_show = [["_" for i in range(3)] for j in range(3)]

    def print_board(self):
        for i in range(3):
            for j in range(3):
                print(self.board_to_show[i][j], end=' ')
            print('\n')

    def move(self, player, id):
        position = input("Выберите клетку('ряд столбик'):\n")
        position = position.split(' ')
        player_move = [int(position[0]) - 1, int(position[1]) - 1]
        self.board_to_show[player_move[0]][player_move[1]] = player.name
        if id == 1:
            self.board[player_move[0]][player_move[1]] = 1  
        else:
            self.board[player_move[0]][player_move[1]] = -1
        self.print_board()

    def trace(self):
        tr = 0
        for i in range(3):
            tr += self.board[i][i]
        return tr

    def anti_trace(self):
        anti_tr = 0
        for i in range(3):
            anti_tr += self.board[i][2 - i]
        return anti_tr

    def column_sum(self, column):
        summ = 0
        for i in range(3):
            summ += self.board[i][column]
        return summ

    def row_sum(self, row):
        summ = 0
        for i in range(3):
            summ += self.board[row][i]
        return summ

    def check_victory(self, player_id):
        if player_id == 1:
            max_col_sum = 0
            max_row_sum = 0
            for i in range(3):
                max_column_sum = max(max_col_sum, self.column_sum(i))
                max_row_sum = max(max_row_sum, self.row_sum(i))
            return max(self.trace(), self.anti_trace(), max_column_sum, max_row_sum)

        if player_id == 2:
            min_col_sum = 3
            min_row_sum = 3
            for i in range(3):
                min_column_sum = min(min_col_sum, self.column_sum(i))
                min_row_sum = min(min_row_sum, self.row_sum(i))
            return min(self.trace(), self.anti_trace(), min_column_sum, min_row_sum)

    def play_game(self):
        steps = 0
        player1 = Player()
        player2 = Player()
        player1.name = input("Игрок 1, ваш id: ")[0]
        player2.name = input("Игрок 2, ваш id:")[0]
        self.print_board()
        while steps != 8:
            print("Игрок 1 ходит: ")
            self.move(player1, 1)
            if self.check_victory(1) == 3:
                print("Игрок 1 победил!")
                return
            print("Игрок 2 ходит: ")
            self.move(player2, 2)
            if self.check_victory(2) == -3:
                print("Игрок 2 победил!")
                return
            steps += 2
        print("Игрок 1 ходит: ")
        self.move(player1, 1)
        if self.check_victory(1) == 3:
            print("Игрок 1 победил!")
            return
        else:
            print("Ничья!")
            return


new_board = Game()
new_board.play_game()
```
